<h3> Importando Biblioteca para Manipulação, Leitura e Análise dos Dados </h3>

```python
import pandas as pd
```

<h3><b> Lendo arquivo .csv </b></h3>

```python
df_ocorrencias_sac = pd.read_csv(
    "ocorrencias_sac_v2.csv", sep=",", encoding="UTF-8",
)

df_ocorrencias_sac
```

<h3><b> Convertendo a coluna <i>'data_ocorrencia'</i> para o formato de data </b></h3>

```python
df_ocorrencias_sac["data_ocorrencia"] = pd.to_datetime(
    df_ocorrencias_sac["data_ocorrencia"], format="%Y-%m-%d"
)
```

<h3><b> Calculando e armazenadno a quantidade de ocorrências por dia</b></h3>

```python
frequencia_ocorrencias = (
    df_ocorrencias_sac["data_ocorrencia"].value_counts().sort_index()
)

ocorrencia_dia = frequencia_ocorrencias.values
```

<h3>Importando Biblioteca para a Criação dos Gráficos</h3>

```python
import matplotlib.pyplot as plt
```

<h3><b> Criando Histograma de Distribuição de Frequência de Ocorrências </b></h3>

```python
plt.figure(figsize=(12, 7))
plt.hist(
    ocorrencia_dia,
    bins=range(min(ocorrencia_dia), max(ocorrencia_dia) + 2),
    edgecolor="black",
    alpha=0.85,
    align="left",
    rwidth=0.9,
)
plt.title(
    "Distribuição de Frequência das Ocorrências do SAC por Dia (2024 - 2025)",
    fontsize=15,
)
plt.xlabel("Número de Ocorrências por Dia")
plt.ylabel("Frequência (Dias)")
plt.xticks(range(int(min(ocorrencia_dia)), int(max(ocorrencia_dia)) + 1))
plt.grid(axis="y")

plt.savefig(
    "distribuicao_frequencia_ocorrencias_sac.pdf",
    transparent=False,
    dpi=600,
    bbox_inches="tight",
)
plt.show()
```
